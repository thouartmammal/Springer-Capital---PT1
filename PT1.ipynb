```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
```

```python
df = pd.read_csv("/home/hle12/Internship/csv files/5.12_Cybersecurity_(detail).csv")
```

```python
# Visualize the average compliance score for each control.
control = df['Control'].unique()
print(control)
dictionary = {}
for control_name in control:
    control_name_rows = df[df['Control'] == control_name]
    dictionary[control_name] = control_name_rows['Compliance_Score'].mean()

plt.bar(dictionary.keys(), dictionary.values(), 0.5, 2)
plt.xlabel('Control')
plt.ylabel('Compliance Score')
plt.xticks(rotation=90)
plt.show()
```
